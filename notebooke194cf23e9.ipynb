```python
import re

# Sample extracted text (replace this with your actual extracted text)
extracted_text = """
Size Comparison
Height
16 cm
Example size
Samsung S9+
6.5 cm
8.5cm
7.5cm
"""

# Function to extract the value after a specific label (e.g., "Height", "Depth", etc.)
def extract_labeled_value(text, label):
    # Pattern to find the label and capture the number followed by 'cm'
    label_pattern = rf'{label}\s*([\d\.]+)\s*cm'
    
    # Search for the pattern in the text
    match = re.search(label_pattern, text, re.IGNORECASE)
    
    # If a match is found, return the corresponding value
    if match:
        return f"{match.group(1)} cm"
    else:
        return f"{label} not found."

# Extract the depth value (assuming we want the "Height" value for depth in this case)
depth_value = extract_labeled_value(extracted_text, "Height")
print(f"Extracted Depth: {depth_value}")
```

```python
import re

def extract_item_weight(text):
    # Clean and convert text
    text = re.sub(r'LBS', 'pound', text)
    
    # Extract weight using regex
    match = re.search(r'(\d+(\.\d+)?)\s?pound', text)
    
    if match:
        weight = match.group(1)
        return f"Item Weight: {weight} pound"
    else:
        return "Item Weight not found"

# Example extracted text
extracted_text = '7" 44 6.75 LBS'

# Extract and print the item weight
item_weight = extract_item_weight(extracted_text)
print(item_weight)
```

```python
!pip install --upgrade keras-ocr tensorflow
import keras_ocr
import os
from PIL import Image
import numpy as np

# Initialize the Keras-OCR pipeline
pipeline = keras_ocr.pipeline.Pipeline()

# Function to extract text from an image using Keras-OCR
def extract_text_from_image_kerasocr(image_path):
    try:
        # Load the image
        img = keras_ocr.tools.read(image_path)
        
        # Perform OCR
        predictions = pipeline.recognize([img])
        
        # Extract text
        text = ' '.join([word[0] for line in predictions[0] for word in line])
        return text
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None

# Example usage
image_dir = 'images'  # Update this path
example_image_path = os.path.join(image_dir, "658003_width.jpg")

extracted_text = extract_text_from_image_kerasocr(example_image_path)
print("Extracted Text using Keras-OCR: ", extracted_text)
```

```python
import os
import pandas as pd
import requests
from PIL import Image
import easyocr

# Initialize the EasyOCR reader
reader = easyocr.Reader(['en'])

# Directory to save downloaded images
download_dir = 'downloaded_images'
os.makedirs(download_dir, exist_ok=True)

# Function to download image from URL
def download_image(url, img_path):
    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(img_path, 'wb') as f:
            f.write(response.content)
        return img_path
    except Exception as e:
        print(f"Error downloading image from {url}: {e}")
        return None

# Function to extract text from an image using EasyOCR
def extract_text_from_image_easyocr(image_path):
    try:
        results = reader.readtext(image_path)
        text = ' '.join([result[1] for result in results])
        return text
    except Exception as e:
        print(f"Error processing image {image_path} with EasyOCR: {e}")
        return None

# Function to process images and metadata together while preserving order
def process_image_metadata(df):
    for index, row in df.iterrows():
        img_name = f"{row['group_id']}_{row['entity_name']}.jpg"
        img_path = os.path.join(download_dir, img_name)

        # Download the image
        downloaded_img_path = download_image(row['image_link'], img_path)
        
        if downloaded_img_path and os.path.exists(downloaded_img_path):
            print(f"Processing {img_name} for group_id {row['group_id']} and entity_name {row['entity_name']}")
            # Extract text using EasyOCR
            extracted_text = extract_text_from_image_easyocr(downloaded_img_path)
            print(f"Extracted Text: {extracted_text}")
        else:
            print(f"Failed to download or process image: {img_name}")

# Example usage with your actual CSV file
df = pd.read_csv('/kaggle/input/sample-test/dataset/sample_test.csv')  # Load your CSV file here

# Call the function
process_image_metadata(df)
```

```python
import pandas as pd
import re

# Define conversion factors for each unit to a common unit
conversion_factors = {
    'centimetre': 1, 'foot': 30.48, 'inch': 2.54, 'metre': 100, 'millimetre': 0.1, 'yard': 91.44,
    'gram': 1, 'kilogram': 1000, 'microgram': 1e-6, 'milligram': 0.001, 'ounce': 28.3495, 'pound': 453.592, 'ton': 1000000,
    'kilovolt': 1000, 'millivolt': 0.001, 'volt': 1,
    'kilowatt': 1000, 'watt': 1,
    'centilitre': 10, 'cubic foot': 28316.8, 'cubic inch': 16.387, 'cup': 237, 'decilitre': 100, 'fluid ounce': 29.5735, 'gallon': 3785.41, 'imperial gallon': 4546.09, 'litre': 1000, 'microlitre': 1e-6, 'millilitre': 1, 'pint': 473.176, 'quart': 946.353
}

# Sample data
data = [
    {'group_id': '658003', 'entity_name': 'width', 'extracted_text': '2OCM 1SCM'},
    {'group_id': '276700', 'entity_name': 'voltage', 'extracted_text': 'Professional tools Blade Diameter Sosti Rated Voltage  2O" Frcuenti JQHZ Rore dinnut Pote 2100W" Ao LoadSceed: 5 JTPIn Capcint 50: HI0 $ Wjeorm Max Cutting Capacity- 60? H6O : WGbran Max Cutting C~Pacily H6o x W2IOrmn Curtne'},
    # Add other data entries here
]

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(data)

# Define the entity to unit mapping
entity_unit_map = {
    'width': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'depth': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'height': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'item_weight': {'gram', 'kilogram', 'microgram', 'milligram', 'ounce', 'pound', 'ton'},
    'maximum_weight_recommendation': {'gram', 'kilogram', 'microgram', 'milligram', 'ounce', 'pound', 'ton'},
    'voltage': {'kilovolt', 'millivolt', 'volt'},
    'wattage': {'kilowatt', 'watt'},
    'item_volume': {'centilitre', 'cubic foot', 'cubic inch', 'cup', 'decilitre', 'fluid ounce', 'gallon', 'imperial gallon', 'litre', 'microlitre', 'millilitre', 'pint', 'quart'}
}

# Extract and convert values
def extract_and_convert_value(entity_name, extracted_text):
    units = entity_unit_map.get(entity_name, set())
    
    # Regular expression to extract numbers and units
    unit_regex = re.compile(r'(\d+\.?\d*)\s*(' + '|'.join(units) + r')', re.IGNORECASE)
    matches = unit_regex.findall(extracted_text)
    
    if matches:
        # Extract the first match and convert
        value, unit = matches[0]
        value = float(value)
        if unit in conversion_factors:
            converted_value = value * conversion_factors[unit]
            return f"{converted_value} {unit}"
    
    # If no match found, return a default message
    return "Value not found"

# Apply the function to each row and create a new 'predicted_value' column
df['predicted_value'] = df.apply(lambda row: extract_and_convert_value(row['entity_name'], row['extracted_text']), axis=1)

# Display the DataFrame
print(df)
```
